1. Porcentaje de empleados ausentes (%) = (Número de empleados que tuvieron al menos 1 ausencia DIVIDIDO Número total de empleados ) x 100

```python
# Calcular total de horas de absentismo anual
total_hores_abs = RRHH['Absenteeism_hours'].sum()

# Calcular total de empleados únicos en el dataset
total_empleados = RRHH['ID'].nunique()

# Filtrar solo empleados con ausencias reales (horas > 0)
empleados_con_ausencia_real = RRHH[RRHH['Absenteeism_hours'] > 0]['ID'].nunique()

# Calcular el porcentaje de empleados ausentes
# ADVERTENCIA: Esto asume que el dataset incluye a todos los empleados de la empresa.
# Si no, el porcentaje no será correcto.
porcentaje_ausentes = (empleados_con_ausencia_real / total_empleados) * 100

# Mostrar los resultados
print("="*50)
print("ESTADÍSTICAS DE ABSENTISMO")
print("="*50)
print(f"Total horas absentismo anual: {total_hores_abs:.2f}")
print(f"Total empleados únicos: {total_empleados}")
print(f"✅ Porcentaje de empleados ausentes: {porcentaje_ausentes:.2f}%")
print(f"   - Empleados totales en dataset: {total_empleados}")
print(f"   - Empleados con ausencias reales: {empleados_con_ausencia_real}")

# Advertencia
if empleados_con_ausencia_real == total_empleados:
    print("\n⚠️  ADVERTENCIA: El porcentaje puede ser 100% porque el dataset solo incluye empleados con ausencias.")
    print("   Si necesitas el porcentaje sobre el total de empleados de la empresa, debes proporcionar el total de empleados.")
else:
    print(f"\nNota: El cálculo asume que el dataset incluye a todos los empleados de la empresa.")
    print(f"      Empleados sin ausencias en el dataset: {total_empleados - empleados_con_ausencia_real}")

print("="*50)
```

```python
# Agrupar per mes i calcular la suma d'hores d'absentisme
absentisme_per_mes = RRHH.groupby('Month_absence')['Absenteeism_hours'].sum()

# Calcular el nombre d'empleats únics per mes
empleats_per_mes = RRHH.groupby('Month_absence')['ID'].nunique()

# Calcular la taxa per càpita mensual
taxa_per_capita = absentisme_per_mes / empleats_per_mes

# Crear un DataFrame amb els resultats mensuals
resultat = pd.DataFrame({
    'Hores_abs': absentisme_per_mes,
    'Num_empleats': empleats_per_mes,
    'Taxa_per_capita': taxa_per_capita
})

# Ordenar els mesos per l'ordre natural
ordre_mesos = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 
               'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
resultat = resultat.reindex(ordre_mesos)

# CALCULS GLOBALS
# Total d'hores d'absentisme anual
total_hores_abs = RRHH['Absenteeism_hours'].sum()

# Nombre total d'empleats únics a l'any
total_empleats_unics = RRHH['ID'].nunique()

# Taxa per càpita global (anual)
taxa_global = total_hores_abs / total_empleats_unics

# Mostrar el resultat mensual
print("Taxa mensual d'absentisme per càpita:")
print(resultat)
print("\n" + "="*50)
print("TAXA GLOBAL D'ABSENTISME PER CÀPITA (ANUAL)")
print(f"Total hores absentisme anual: {total_hores_abs:.2f} hores")
print(f"Total empleats únics: {total_empleats_unics} empleats")
print(f"Taxa per càpita global: {taxa_global:.2f} hores/empleat")
print("="*50)

# Exportar a CSV (amb una fila addicional per la global)
resultat_csv = resultat.copy()
resultat_csv.loc['Global'] = [total_hores_abs, total_empleats_unics, taxa_global]
resultat_csv.to_csv('taxa_absentisme_per_capita_mensual.csv', index=True)

# Opcional: Crear un DataFrame separat per les estadístiques globals
global_stats = pd.DataFrame({
    'Metrica': ['Total hores absentisme', 'Total empleats', 'Taxa per càpita global'],
    'Valor': [total_hores_abs, total_empleats_unics, taxa_global]
})
global_stats.to_csv('estadistiques_globals_absentisme.csv', index=False)
```

2. Mes pico de carga por ausencias:
El mes con la maxima suma de horas de absentismo

```python
# Agrupar por mes y sumar las horas de absentismo
absentismo_por_mes = RRHH.groupby('Month_absence')['Absenteeism_hours'].sum()

# Encontrar el mes con el máximo valor
mes_pico = absentismo_por_mes.idxmax()
max_horas = absentismo_por_mes.max()

print(f"El mes pico de carga por ausencias es {mes_pico} con {max_horas} horas de absentismo.")
```

3. Motivo modal de absentismo laboral:
Maxima frequencia de motivo de ausencia


```python

# Calcular el motivo modal de absentismo
motivo_modal = RRHH['Reason_absence'].mode()[0]
frecuencia_modal = RRHH['Reason_absence'].value_counts().max()

# Obtener la descripción del motivo modal del diccionario
descripcion_modal = reason_absence_dict.get(motivo_modal, "Motivo desconocido")

# Imprimir el resultado
print(f"El motivo modal de absentismo es: {descripcion_modal} (Código: {motivo_modal})")
print(f"Frecuencia: {frecuencia_modal} ausencias")
```
